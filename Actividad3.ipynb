CONCEPTOS BÁSICOS DE LA LIBRERÍA MATPLOTLIP.
Matplotlib es una librería de trazado de gráficos de bajo nivel en Python, que sirve como una herramienta de visualización. Matplotlib es de código abierto y podemos utilizarlo libremente. Matplotlib está escrito principalmente en Python por algunos segmentos escritos con C.
INSTALACIÓN DE MATPLOTLIB
Si tienes Python y pip  ya instalado en tu sistema instala Matplotlib usando el siguiente comando:



```python
pip install matplotlib
```

Desde los comandos importar Matplotlib una vez instalado Matplotlib impórtalo en tus aplicaciones, agregando el módulo import, ejemplo

```python
import matplotlib
```

MATPLOTLIB PYPLOT
Pyplot la mayoría de las herramientas de Matplotlib se encuentran dentro del submódulo pyplot y normalmente se importa bajo el alias PLT:

```python
import matplotlib.pyplot as plt
```

Ahora el paquete Pyplot puede denominarse plt.
